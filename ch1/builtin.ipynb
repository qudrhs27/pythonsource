```python
# abs(x) : 절대값

print(abs(3))
print(abs(-3))
print(abs(312))
```

```python
# chr() <=> ord()
# chr() : 유니코드 값을 넣으면 그 코드에 해당하는 문자를 리턴
# ord() : 문자를 넣으면 문자에 해당하는 유니코드를 리턴

print(chr(97))
print(chr(44032))

print(ord('a'))
print(ord('가'))
```

```python
# dir() : 객체가 지닌 변수나 함수를 보여주는 함수

list1 = [1,2,3]
dir(list1)

import math
dir(math)
```

```python
# divmod(a, b) : 몫,나머지를 튜플로 반환
quot, rem = divmod(7, 3)
print(quot)
print(rem)
```

```python
# enumerate() : 리스트, 튜플, 문자열을 입력으로 받아 인덱스 값을 포함하는 enumerate 객체 반환

for i, name in enumerate(['body', 'foo', 'bar'], start=1):
    print(i, name)

for i, name in enumerate('bodybar', start=1):
    print(i, name)
```

```python
print(max([1,2,3]))
print(max("python"))
print(min([1,2,3]))
print(min("python"))
```

```python
# round(숫자 [,ndigits]) : 반올림

print(round(4.6))
print(round(4.678, 2))
```

```python
# sorted() : 입력 데이터를 정렬 후 결과를 새로운 리스트로 리턴
list1 = [1,3,2,6,5]
# 원본 정렬
# list1.sort()

print(sorted(list1))
print(list1)
```

```python
# zip() : 동일한 개수로 이루어진 데이터들을 묶어서 반환해주는 함수

print(list(zip([1,2,3],[4,5,6])))
print(list(zip([1,2,3],[4,5,6],[7,8,9])))
print(list(zip("abc","def")))
```

```python
# 학생 5명의 점수를 입력받아 다음 정보를 출력하는 프로그램을 작성

# 출력내용
# 최고점수(max), 최저점수(min), 평균점수(소수 둘째 자리)(sum,len), 점수를 오름차순으로 출력(sorted), 등수와 함께 출력(enumerate)

score = [int(input("점수를 입력하세요 : ")) for _ in range(5)]

print(f"최고점수 : {max(score)}")
print(f"최저점수 : {min(score)}")
print(f"평균점수 : {sum(score)/len(score)}")
print(f"점수 오름차순으로 출력 : {sorted(score)}")

for idx, score in enumerate(sorted(score, reverse=True), start=1):
    print(f"{idx}등 : {score}")
```

```python
# 문자열을 입력받아 다음 정보를 출력
# 출력내용 : 입력한 문자, 각 문자의 유니코드 값, 가장 큰 문자, 가장 작은 문자

# ord(), chr(), max(), min(), list()
# 문자열 입력 : Python
# P -> 80
# y -> 121
#
# 가장 큰 문자 : y
# 가장 작은 문자 : P

string = input("문자열 입력 : ")
print(f"입력한 문자 : {string}")

for word in string:
    print(f"{word} -> {ord(word)}")

print(f"가장 큰 문자 : {max(string)}")
print(f"가장 작은 문자 : {min(string)}")
```

```python
x = [3,7,1,-5,9]
y = [2,-4,8,10,-1]

# zip(), abs(), enumerate()
# 출력
# (3,2) -> 5

for idx, (a, b) in enumerate(zip(x,y), start=1):
    distance = abs(a) + abs(b)
    print(f"{idx}번 좌표 : ({a}, {b}) -> {distance}")
```

```python
# 사용자로부터 숫자 10개 입력받기

# 출력
# 입력리스트, 중복제거, 오름차순, 튜플변환, 집합 크기

# 3 5 2 5 7 ~~

# 원본 [3,5,2,5,7~~]
# 중복제거 {1,2,3,4} set()
# 정렬 sorted()
# 튜플 tuple()
# 집합 크기 : len()

num = [int(input("숫자 입력")) for _ in range(10)]

print(f"원본 리스트 : {num}")
unique_numbers = list(set(num))
print(f"중복제거 : {unique_numbers}")
sorted_numbers = sorted(unique_numbers)
print(f"정렬 : {sorted_numbers}")
print(f"튜플 : {tuple(sorted_numbers)}")
print(f"집합 크기 : {len(unique_numbers)}")
```
