```python
# 결과저장 변수
result1 = 0
result2 = 0

def add(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add(3))
print(add(4))

print(add2(13))
print(add2(14))
```

#### 클래스
- 설계도
- 클래스명은 대문자로 시작
- 메소드(==함수)

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

    def sub(self, num):
        self.result -= num
        return self.result
```

```python
# 객체 생성 => __init__() 자동으로 호출됨 
# 변수 = 클래스명()

calc1 = Calculator()
print(calc1.add(3))
print(calc1.add(4))


calc2 = Calculator()
print(calc2.add(13))
print(calc2.add(14))
```

```python
class Student:
    def __init__(self, name, number, grade, detail):
        self.name = name
        self.number = number
        self.grade = grade
        self.detail = detail

    # 인스턴스를 출력할 때 원하는 값들을 반환
    def __str__(self):
        return f"이름 : {self.name}, 반 : {self.number}, 학년 : {self.grade}, 상세정보 : {self.detail}"
```

```python
# 객체 생성 => __init__() 호출
s1 = Student("kim", 1, 1, {'gender':'male', 'score1':95, 'score':88})
s2 = Student("Park", 2, 2, {'gender': 'female', 'score1': 75, 'score': 68})
s3 = Student("Hong", 2, 1, {'gender': 'male', 'score1': 99, 'score': 86})

# s1은 Student 클래스의 인스턴스임
print(s1) # <__main__.Student object at 0x00000252881AA9E0>
print(s2) # <__main__.Student object at 0x00000252881AA8B0>
print(s3) # <__main__.Student object at 0x0000025288355370>
```
