````python
# 한 시간 임금은 10700원, 8시간 까지는 10700으로 지급하고 8시간 초과 근무시에는 1.5배 지금

# 근무 시간(hours), rate(시간당 임금), pay(일당)
hours, rate, pay = 12, 10700, 0
if hours <= 8:
    pay = hours * rate
else:
    pay = rate * 8 + (10700 * 1.5) * (hours - 8)

print("근무시간 : %d, 일당 : %d" % (hours, pay))
print("근무시간 : {0}, 일당 : {1:.0f}".format(hours, pay))
print(f"근무시간 : {hours}, 일당 : {pay:.0f}")
````

````python
# 주차장 프로그램

# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A']
# 1
# B 자동차 입차. 주차장 상태 ==> ['A', 'B']
# 2
# B 자동차 출차. 주차장 상태 ==> ['A']
# E 자동차 입차. 주차장 상태 ==> ['A', 'B', 'C', 'D', 'E']
# 1
# 주차장 만차


parking_lot = []
top, car_name = 0, 'A'

while True:
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료 : "))

    if no <= 3:
        if no == 1:
            # top >= 5 주차장 만차 => 아무것도 안함

            # A 자동차 입차. 주차장 상태 ==> ['A']
            # parking_lot.append(car_name)
            # car_name = chr(ord(car_name)+1)
            # top = top + 1
            if top >= 5:
                print("주차장 만차")
            else:
                parking_lot.append(car_name)
                print(f"{car_name} 자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1)
                top = top + 1
        elif no == 2:
            # top > 0
            # 리스트 부터 제거 : pop()
            # B 자동차 출차. 주차장 상태 ==> ['A']
            # top = top - 1, car_name-1
            
            # 빠져나갈 자동차 없음
            if top > 0:
                outCar = parking_lot.pop()
                print(f"{outCar} 자동차 출차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)-1)
                top = top - 1
            else:
                print("빠져나갈 자동차 없음")
        else:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인해 주세요")

    
````
