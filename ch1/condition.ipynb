```python
num = 5

# if 문 : bool 자료형
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70 이상이면 C, 나머지는 F

score = 99

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
```

```python
# a = 12
# b = 6
# c = 19

a, b ,c = 12, 6, 19

if a > b and a > c:
    print(a)
elif b > a and b > c:
    print(b)
else:
    print(c)
```

```python
fruits = ['딸기', '사과', '바나나', '배', '수박']

# fruits 리스트에 딸기가 포함되어 있다면 딸기가 포함됨, 없다면 딸기가 없음
# in, if

if '딸기' in fruits:
    print("딸기가 포함됨")
else:
    print("딸기가 없음")
```

```python
fruits.clear()

if fruits:
    print("리스트")
else:
    print("빈 리스트")


if not fruits:
    print("빈 리스트")


# if len(fruits) == 0:
#     print("빈 리스트")
```

```python
# and : x and y(x,y가 모두 참인 경우 참)
# or : x or y (x,y 둘 중 하나만 참이면 참)
# not x

# score가 90 이상이고, grade가 A인 경우 합격, 나머지 불합격

score, grade = 85, "A"
if score >= 90 and grade == "A":
    print("합격")
else:
    print("불합격")
```

```python
# score가 90 이상이거나, grade가 A인 경우 합격, 나머지 불합격

score, grade = 85, "A"
if score >= 90 or grade == "A":
    print("합격")
else:
    print("불합격")
```
