#### file 읽고 쓰기

| 파일 열기 모드 | 설명 | 
| ----------- | ----- |
| `r` | **읽기(Read) 모드 ** : 파일의 내용을 읽기만 할 때 사용. 파일이 존재하지 않는 경우 오류 발생. |
| `w` | **쓰기(Write) 모드 ** : 파일의 내용을 쓰기만 할 때 사용. 기존 파일이 존재하는 경우 내용 삭제 후 새로 작성. 없는 경우 새로 생성. |
| `a` | **추가(Append) 모드 ** : 기존 파일의 마지막에 내용을 추가할 때 사용. 파일이 없는 경우 새로 생성. |

```python
# score.txt 내용을 읽은 후 평균 구한 후 출력

scores = []
with open("./data/score.txt", "r", encoding="utf-8") as f:
    for num in f:
        scores.append(int(num))

print(scores)
print(f"평균: {sum(scores) / len(scores)}")
```

```python
# sample.txt 파일을 읽어서 총합과 평균을 구한 후 총합과 평균을 result.txt 파일 작성

scores = []
with open("./data/sample.txt", "r", encoding="utf-8") as f:
     for num in f:
        scores.append(int(num))

with open("./data/result.txt", "w", encoding="utf-8") as f:
    f.write(f"총합: {sum(scores)}\n평균: {sum(scores) / len(scores)}")
```

```python
# fruits.txt 읽어서 단어의 개수 출력
# apple 3
# banana 2

# apple banana orange
# 단어 분리 : split() => ['apple', 'banana', 'orange']

# dict 추가
# if "apple" in dict1:
#     dict1['apple'] += 1
# else:
#     dict1['apple'] = 1

dict1 = {}
# with open("./data/fruits.txt", "r", encoding="utf-8") as f:
#     list1 = f.read().split()
#     for i in range(len(list1)):
#         if list1[i] in dict1:
#             dict1[list1[i]] += 1
#         else:
#             dict1[list1[i]] = 1

with open("./data/fruits.txt", "r", encoding="utf-8") as f:
    for line in f:
        words = line.split()
        for word in words:
            if word in dict1:
                dict1[word] += 1
            else:
                dict1[word] = 1

for fruit,count in dict1.items():
    print(f"{fruit} : {count}")
```

```python
# # 1000명의 이름, 키, 몸무게 info.txt 읽은 후
# # BMI 지수 계산
# bmi = 체중 / 키의 제곱
# bmi가 18.5 미만일 때 저체중
# bmi가 18.5 ~ 22.9 미만일 때 정상체중
# bmi가 23 ~ 24.9 미만일 때 과체중
# bmi가 25 이상 비만

# 이름 : 하타
# 몸무게 : 95
# 키 : 180
# BMI : 18.6
# 결과 : 정상체중

with open("./data/info.txt", "r", encoding="utf-8") as f:
    for line in f:
        name,weight,height = line.strip().split(", ")
        bmi = int(weight) / (int(height) * 0.01) ** 2

        if bmi < 18.5:
            result = "저체중"
        elif 18.5 < bmi < 22.9:
            result = "정상체중"
        elif 23 < bmi < 24.9:
            result = "과체중"
        else:
            result = "비만"

        print("\n".join(["이름 : {}", "몸무게 : {}", "키 : {}", "BMI : {:.1f}", "결과 : {}"]).format(name,weight,height,bmi,result))
        print()
```

```python
# origin.txt 읽어서 암호화 시킨 후 encr.txt 저장
# 1. 암호화 선택 : origin.txt 읽어서 변수에 담기 => 그걸 chr(ord("안")+100) 저장
# 2. 복호화 선택 : 암호화된 파일을 읽은 후 원래대로 출력

while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 중 원하는 메뉴의 번호를 입력하세요"))

    if no == 1:
        with open("./data/origin.txt", "r", encoding="utf-8") as f:
            contents = f.read()
        
        with open("./data/encr.txt", "w", encoding="utf-8") as f:
            for c in contents:
                f.write(chr(ord(c)+100))
    elif no == 2:
        with open("./data/encr.txt", "r", encoding="utf-8") as f:
            contents = f.read()
            for c in range(len(contents)):
                print(chr(ord(contents[c]) - 100), end="")
    else:
        break
```
