#### 정규표현식(정규식)
- 복잡한 문자열을 처리할 때 사용하는 방법
- 파이썬만의 고유한 문법이 아니라 문자열을 처리하는 모든 곳에서 사용하는 일종의 형식 언어

```python
# 주민등록번호 -> 뒷자리를 모두 * 문자로 변경

# 파이썬에서 정규식 지원하는 모듈
import re

dir(re)

# re.compile() : 정규식을 작성

pattern = re.compile("ab*")

# 패턴을 특정 문자열이 포함하고 있는가?
print(pattern.match("abc"))
```

| 메서드 | 목적 |
| ---- | ---- |
| match | 문자열의 처음부터 정규식과 매치되는지 조사 |
| search | 문자열 전체를 검색하여 정규식과 매치되는지 조사 |
| findall | 정규식과 매치되는 모든 문자열(substring)을 리스트로 리턴 |
| finditer | 정규식과 매치되는 모든 문자열(substring)을 이터레이더 객체로 리턴 |

```python
# [문자] : [] 안에 들어있는 문자들과 매치하느냐?

p = re.compile("[abc]")

print(p.match("a"))
print(p.match("before"))
print(p.match("dude"))
```

```python
m = p.match("a")

print(f"패턴 시작 : {m.start()}")
print(f"패턴 끝 : {m.end()}")
print(f"패턴과 일치하는 문자열 반환 : {m.group()}")
print(f"일치의 시작,끝 반환 : {m.span()}")

if m:
    print(f"Match found {m.group()}")
else:
    print("No Match")
```

```python
p = re.compile("[abc]")

m = p.findall("Happy forever")
print(m)

m = p.finditer("Happy forever")
print(m)
for r in m:
    print(r)
```

```python
# . 문자 : \n을 제외한 모든 문자

p = re.compile("a.b")

print(p.match("a\nb"))
print(p.match("abb"))
print(p.match("a0b"))
print(p.match("aab"))
print(p.match("a b"))
```

```python
# . : 마침표로서 사용하고 싶을 때
# \. OR [.]
# \t,\r...
p = re.compile(r"a\.b")
# p = re.compile(r"a[.]b")

print(p.match("a\nb"))
print(p.match("abb"))
print(p.match("a0b"))
print(p.match("aab"))
print(p.match("a b"))
print(p.match("a.b"))
```

- 자주 사용하는 문자 표현식

| 정규표현식 | 설명 |
| ---- | ---- |
| \d | 숫자와 매치 [0-9]와 동일한 표현식 |
| \D | 숫자가 아닌 것과 매치 [^0-9]와 동일한 표현식 |
| \s | 화이트스페이스(스페이스,탭,줄바꿈 등) 문자와 매치 [빈칸\t\n\r\f\v] 와 동일한 표현식 |
| \S | 화이트스페이스(스페이스,탭,줄바꿈 등) 문자가 아닌 것과 매치 [^빈칸\t\n\r\f\v] 와 동일한 표현식 |
| \w | 문자 + 숫자와 매치 [a-zA-Z0-9_] 와 동일한 표현식 |
| \W | 문자 + 숫자가 아닌 것과 매치 [^a-zA-Z0-9_] 와 동일한 표현식 |

```python
p = re.compile("[0-9]+")
p = re.compile(r"\d+")
print(p.match("012 abc"))

# 원본 문자열에서 0~9가 아닌
# p = re.compile("[^0-9]+")
p = re.compile(r"\D+")
print(p.match("012 abc"))

# 공백,탭,엔터 문자 찾기
p = re.compile("[\t\n\r\f\v]+")
p = re.compile(r"\s+")
print(p.search("012 abc"))

# 공백,탭,엔터 문자가 아닌 문자 찾기
# p = re.compile("[^ \t\n\r\f\v]+")
p = re.compile(r"\S+")
print(p.search("012 abc"))

# 대소문자,숫자,_ 포함한 문자
# p = re.compile("[a-zA-Z0-9_]+")
p = re.compile(r"\w+")
print(p.search("012 abc"))

# 대소문자,숫자,_ 가 아닌 문자 찾기
# p = re.compile("[^a-zA-Z0-9_]+")
p = re.compile(r"\W+")
print(p.search("012 abc"))
```

```python
# re.I : ignorecase

# 영어가 포함되어 있는지?
p = re.compile("[a-z]+", re.I)
print(p.match("python"))
print(p.match("Python"))
print(p.match("PYTHON"))
```

```python
# re.M : MULTILINE

data = """python one
life is too short
python two
you need python
python three
"""

# ^ : 시작하는
# python 시작하고 스페이스문자하나 영어,숫자,_ 인 문자를 찾아줘
p = re.compile(r"^python\s\w+", re.M)
print(p.findall(data))
```

```python
# | : or

p = re.compile("Crow|Servo")
m = p.match("CrowHello")
print(m)

m = p.match("ServoHello")
print(m)
```

```python
# ^ : 시작하는,   $ : 끝나는

p = re.compile("^Life")
m = p.search("Life is too short")
print(m)

m = p.search("My Life")
print(m)

p = re.compile("short$")
m = p.search("Life is too short")
print(m)
```

```python
# () : 그루핑

# ABCABC 문자열이 반복되는지 조사

p = re.compile("[ABC]+")
m = p.search("ABCABCABC OK?")
print(m)
print(m.group())
```

```python
p = re.compile("(ABC)+")
m = p.search("ABCABCABC OK?")
print(m)
print(m.group())
print(m.group(1))
```

```python
# 이름 010-1234-5678
p = re.compile(r"(\w+)\s+(\d{3}-\d{4}-\d{4})")
m = p.search("park 010-1234-5678")
print(m)
print(m.group())
print(m.group(1))
print(m.group(2))
```

```python
# 찾아서 바꾸기
# John Smith
# Smith, John

p = re.compile(r"(\w+)\s(\w+)")
m = p.search("John Smith")
print(m)

# re.sub(패턴, 바꿀 문자열, 원본 문자열)
re.sub(p, r"\2, \1", "John Smith")
```

```python
# re.compile()
# re.sub()

# re.fullmatch() : 문자열 전체가 패턴과 일치

p = re.compile(r"\d+")

print(re.fullmatch(p, "1232abc"))
print(re.fullmatch(p, "123456"))
print(re.fullmatch(p, "abc123"))
```

```python
# 영문 소문자로 시작하고 영문,숫자 포함된 5~11글자

p = re.compile("[a-z][a-z0-9]{5,11}")
print(re.fullmatch(p, "1232abc"))
print(re.fullmatch(p, "123456"))
print(re.fullmatch(p, "abc123"))

m = re.fullmatch(p, "abc123")
print(m.group())
```

```python
# 입력을 받은 값이 숫자인지 확인

# 숫자입니다 or 숫자가 아닙니다 출력
# fullmatch()

num = input("숫자를 입력하세요 : ")
p = re.compile(r"\d+")

if re.fullmatch(p, num):
    print("숫자입니다")
else:
    print("숫자가 아닙니다")
```

```python
# 입력값이 영어 대소문자인지 확인하기

text = input("문자를 입력하세요 : ")
p = re.compile("[a-zA-Z]+")

if re.fullmatch(p, text):
    print("영어입니다")
else:
    print("영어가 아닙니다")
```

```python
# 아이디,비밀번호,전화번호 입력 받기
# 패턴에 맞을 때까지 입력받기

# 아이디 검사 조건
# 영문 대문자로 시작, 영문,숫자,특수문자(!@#$&) 포함, 6~15자

# 비밀번호 검사 조건
# 영문 대소문자,숫자,특수문자(!@#$%) 포함, 8자 이상

# 전화번호 010-1234-5678 / 01012345678
# fullmatch()

while True:
    p_id = re.compile("[A-Z][a-zA-Z0-9!@#$&]{6,15}")

    user_id = input("아이디를 입력하세요 : ")
    if re.fullmatch(p_id, user_id):
        print("아이디 입력 완료")
        break
    else:
        print("아이디를 다시 입력하세요.")

while True:
    p_pwd = re.compile("[a-zA-Z0-9!@#$%]{8,}")

    user_password = input("비밀번호를 입력하세요 : ")
    if re.fullmatch(p_pwd, user_password):
        print("비밀번호 입력 완료")
        break
    else:
        print("비밀번호를 다시 입력하세요.")

while True:
    p_tel = re.compile(r"\d{3}-\d{4}-\d{4}")

    user_tel = input("전화번호를 입력하세요 : ")
    if re.fullmatch(p_tel, user_tel):
        print("전화번호 입력 완료")
        break
    else:
        print("전화번호를 다시 입력하세요.")

print(f"아이디 : {user_id}\n비밀번호 : {user_password}\n전화번호 : {user_tel}")
```
